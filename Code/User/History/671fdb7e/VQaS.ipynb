# Difference between Kernel and Max pooling layer

A kernel is the essential foundation of a computer's operating system. <br>
It is the core that provides basic services for all the other parts of the OS

However, in terms of deep learning, the pooling operation involves sliding a two-dimensional window over the input and taking the maximum, average, or sum of the values in the window. <br>

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np

# Generate a random 5x5 image as an example
image = torch.rand(1, 1, 5, 5)  # shape: (batch_size, channels, height, width)

# Define a 3x3 kernel for convolution
kernel = torch.tensor([[1.0, 0.0, -1.0],
                       [1.0, 0.0, -1.0],
                       [1.0, 0.0, -1.0]]).reshape(1, 1, 3, 3)  # shape: (out_channels, in_channels, height, width)

# Apply convolution with the kernel
convolution_output = F.conv2d(image, kernel, stride=1, padding=1)

# Apply max pooling with a 2x2 kernel
max_pool_output = F.max_pool2d(image, kernel_size=2, stride=2)

# Convert tensors to numpy arrays for visualization
image_np = image.squeeze().numpy()
conv_np = convolution_output.squeeze().detach().numpy()
max_pool_np = max_pool_output.squeeze().detach().numpy()

# Plot the original image, convolution output, and max pooling output
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(image_np, cmap='gray')
axes[0].set_title("Original Image")
axes[1].imshow(conv_np, cmap='gray')
axes[1].set_title("After Convolution (Kernel Layer)")
axes[2].imshow(max_pool_np, cmap='gray')
axes[2].set_title("After Max Pooling Layer")

for ax in axes:
    ax.axis('off')

plt.show()
```
